```python
!sort -k1,1 ENCFF111HUS.bed > ENCFF111HUS_sorted.bed
!sort -k1,1 ENCFF143HTO.bed > ENCFF143HTO_sorted.bed
```

```python
!bedtools genomecov -bg -i ENCFF111HUS_sorted.bed -g hg38.chrom.sizes > ENCFF111HUS.bedGraph
!bedtools genomecov -bg -i ENCFF143HTO_sorted.bed -g hg38.chrom.sizes > ENCFF143HTO.bedGraph
```

```python
# download script
!wget -q http://hgdownload.soe.ucsc.edu/admin/exe/linux.x86_64/bedGraphToBigWig
!chmod 777 bedGraphToBigWig
```

```python
!./bedGraphToBigWig ENCFF111HUS.bedGraph hg38.chrom.sizes ENCFF111HUS.bigWig
!./bedGraphToBigWig ENCFF143HTO.bedGraph hg38.chrom.sizes ENCFF143HTO.bigWig
```

## Графики

```python
!wget https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/genes/hg38.knownGene.gtf.gz
```

```python
# подсчёт матрицы ENCFF111HUS
%%time
!computeMatrix scale-regions \
  -S ENCFF111HUS.bigWig \
  -R hg38.knownGene.gtf.gz \
  -out matrix.tab.gz \
  -a 2000 -b 2000 \
  --regionBodyLength 4000 \
  --skipZeros \
  --missingDataAsZero
```

```python
!plotHeatmap \
  -m matrix.tab.gz \
  -out result.png \
  --colorMap YlGnBu \
  --regionsLabel 'genes' \
  --heatmapHeight 15 \
  --plotTitle 'ngs_plot' &
```
